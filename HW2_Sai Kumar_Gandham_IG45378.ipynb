#### 4. ReadinthetextsoftheStateoftheUnionaddresses,usingthestate_unioncorpus reader. Count occurrences of men, women, and people in each document. What has happened to the usage of these words over time?

```python
from nltk.corpus import state_union

nltk.download('state_union')

# Here first define the words to count
words_to_count = ['men', 'women', 'people']

# Let's Create a dictionary to store word counts by year
word_counts_by_year = {}

# the best thing here we can do is iterating through each document
for fileid in state_union.fileids():
    year = fileid[:4]  # Extract the year from the fileid
    words = state_union.words(fileid)  # Get the words from the document
    word_counts = {word: words.count(word) for word in words_to_count}  # Count occurrences of target words
    word_counts_by_year[year] = word_counts  # Store word counts for the year

# Ok, now let's Print word counts by year
print("Word counts by year:")
for year, counts in sorted(word_counts_by_year.items()):
    print(year, counts)
```

```python
#For better understanding let's visualize the above data 

import matplotlib.pyplot as plt

# Extract data for plotting
years = sorted(word_counts_by_year.keys())
word_counts = {word: [word_counts_by_year[year][word] for year in years] for word in words_to_count}

# Plotting
plt.figure(figsize=(10, 6))
for word in words_to_count:
    plt.plot(years, word_counts[word], label=word)
plt.title('Word Counts by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Talking about the word "men," we noticed that its use changes from year to year. Sometimes, it shows up a lot in the State of the Union speeches, and other times, not so much.

For "women," it's similar but different. At first, it's not mentioned much, but over time, it gets mentioned more often.

Now, "people" is interesting. It's used regularly every year, more than "men" or "women".

Overall, when we look at all this, we see that mentions of both "men" and "women" seem to be increasing as time goes on, with some ups and downs. "People" is always there, though, being talked about a lot in these speeches.

#### 5. Investigate the holonym-meronym relations for some nouns. Remember that there are three kinds of holonym-meronym relation, so you need to use member_mer onyms(), part_meronyms(), substance_meronyms(), member_holonyms(), part_holonyms(), and substance_holonyms().

```python
from nltk.corpus import wordnet as wn

# Let's create a function to collect meronyms and holonyms relations for a given noun
def collect_relations(noun):
    synsets = wn.synsets(noun, pos=wn.NOUN)
    meronyms_holonyms = {
        "member_meronyms": [],
        "part_meronyms": [],
        "substance_meronyms": [],
        "member_holonyms": [],
        "part_holonyms": [],
        "substance_holonyms": []
    }
    for synset in synsets:
        meronyms_holonyms["member_meronyms"].extend(synset.member_meronyms())
        meronyms_holonyms["part_meronyms"].extend(synset.part_meronyms())
        meronyms_holonyms["substance_meronyms"].extend(synset.substance_meronyms())
        meronyms_holonyms["member_holonyms"].extend(synset.member_holonyms())
        meronyms_holonyms["part_holonyms"].extend(synset.part_holonyms())
        meronyms_holonyms["substance_holonyms"].extend(synset.substance_holonyms())
        
    # I think it's better to convert synsets to their respective names for better readability
    for relation in meronyms_holonyms:
        meronyms_holonyms[relation] = [meronym.name().split('.')[0] for meronym in meronyms_holonyms[relation]]
        # Remove duplicates
        meronyms_holonyms[relation] = list(set(meronyms_holonyms[relation]))
    return meronyms_holonyms

# Define the nouns to check
nouns = ["car", "people", "tree"]

# Collect and print meronyms and holonyms relations for each noun
for noun in nouns:
    print("Noun:", noun)
    relations = collect_relations(noun)
    print("Member Meronyms:", relations["member_meronyms"])
    print("Part Meronyms:", relations["part_meronyms"])
    print("Substance Meronyms:", relations["substance_meronyms"])
    print("Member Holonyms:", relations["member_holonyms"])
    print("Part Holonyms:", relations["part_holonyms"])
    print("Substance Holonyms:", relations["substance_holonyms"])
    print()
```

#### 17. Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.

```python
from nltk.corpus import stopwords
from collections import Counter

def top_non_stopwords(text):
    # Tokenize the text
    words = nltk.word_tokenize(text.lower())

    # Filter out stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word.isalnum() and word not in stop_words]

    # Count the occurrences of each word
    word_counts = Counter(words)

    # Find the 50 most common words
    top_50_words = word_counts.most_common(50)
    
    return top_50_words

# Let's see an example to check how the above created function is working:
text = "This is a sample text. It contains some words, but not all words are important."
top_words = top_non_stopwords(text)
print("50 Most Frequently occurring Non-Stopwords:")
for word, count in top_words:
    print(word, "-", count)
```

#### 23. Zipfs Law: Let f(w) be the frequency of a word w in free text. Suppose that all the words of a text are ranked according to their frequency, with the most frequent word first. Zipfs Law states that the frequency of a word type is inversely proportional to its rank (i.e., f  r = k, for some constant k). For example, the 50th most common word type should occur three times as frequently as the 150th most common word type.

#### a. Write a function to process a large text and plot word frequency against word rank using pylab.plot. Do you confirm Zipfs law? (Hint: it helps to use a logarithmic scale.) What is going on at the extreme ends of the plotted line?

```python
from nltk.corpus import gutenberg
import pylab

def plot_zipf_law(text):
    #Let's follow our question instructions,first we need to tokenize the text
    tokens = nltk.word_tokenize(text)
    
    # Calculate word frequencies
    freq_dist = nltk.FreqDist(tokens)
    
    # Sort the frequencies in descending order
    sorted_freq = sorted(freq_dist.values(), reverse=True)
    
    # Plot word frequency against word rank
    ranks = range(1, len(sorted_freq) + 1)
    pylab.plot(ranks, sorted_freq, marker='.', linestyle='-')
    pylab.xscale('log')
    pylab.yscale('log')
    pylab.title("Zipf's Law: Word Frequency vs. Word Rank")
    pylab.xlabel("Word Rank (log scale)")
    pylab.ylabel("Word Frequency (log scale)")
    pylab.grid(True)
    pylab.show()

# Example:
large_text = nltk.corpus.gutenberg.raw('melville-moby_dick.txt')
plot_zipf_law(large_text)
```

The diagonal pattern you see at the ends of the graph is because of Zipf's Law, which says that as the frequency of a word decreases, its rank increases. Ideally, when plotted on a logarithmic scale, this relationship should form a straight line. However, in real-world text like "Moby Dick," there are deviations from this ideal line due to different factors.

Despite these deviations, the overall diagonal pattern supports Zipf's Law, showing that word frequency and rank generally have an inverse relationship. This means that the most common words appear much more frequently than less common ones.

#### a. Store the n most likely words in a list words, then randomly choose a word from the list using random.choice(). (You will need to import random first.) :

```python
import random

def generate_model(cfdist, word, num=200):
    words = [word]
    for i in range(num):
        # Append the current word to the list
        words.append(word)
        # Get the n most likely words after the given word
        most_common_words = cfdist[word].most_common(num)
        # Extract just the words
        likely_words = [w for w, _ in most_common_words]
        # Randomly choose the next word from the likely words
        word = random.choice(likely_words)
    print(' '.join(words))

# Let's see an example
text = nltk.corpus.genesis.words('english-kjv.txt')
bigrams = nltk.bigrams(text)
cfd = nltk.ConditionalFreqDist(bigrams)

generate_model(cfd, 'living')
```

#### 25. Define a function find_language() that takes a string as its argument and returns a list of languages that have that string as a word. Use the udhr corpus and limit your searches to files in the Latin-1 encoding.

```python
import nltk

def find_language(word):
    languages = []
    for lang in nltk.corpus.udhr.fileids():
        if lang.endswith('Latin1') and word in nltk.corpus.udhr.words(lang):
            languages.append(lang)
    return languages

# Lets see an example
word = 'people'
languages_with_word = find_language(word)
print("Languages that have '{}' as a word:".format(word))
print(languages_with_word)
```
